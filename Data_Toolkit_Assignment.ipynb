1. What is NumPy, and why is it widely used in Python ?

NumPy (Numerical Python) is a core Python library for numerical and scientific computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-performance mathematical functions to operate on them.


Why NumPy is Widely Used in Python ?

a. Efficient Array Operations:

NumPy’s ndarray is faster and more memory-efficient than Python lists, especially for large datasets.

b. High Performance:

Many NumPy operations are implemented in C and Fortran, making computations much faster than pure Python loops.

c. Mathematical & Statistical Functions:

It includes built-in functions for linear algebra, statistics, Fourier transforms, and more.

d. Vectorization:

Allows performing operations on entire arrays at once without using explicit loops, making code shorter and faster.

e. Foundation for Other Libraries:

Popular libraries like Pandas, SciPy, Matplotlib, TensorFlow, and scikit-learn are built on top of NumPy.

f. Easy Integration:

Works well with C/C++ and other languages, making it suitable for high-performance applications.

2. How does broadcasting work in NumPy?

When two arrays are used in an operation, NumPy compares their shapes from right to left and checks these rules:

a. Dimensions are compatible if

>They are equal, or

>One of them is 1

b. If dimensions are compatible,

>NumPy stretches (broadcasts) the array with dimension 1 to match the other array.

c. If not compatible,

>NumPy raises a ValueError.

17. Explain the difference between apply() and map() in PandasA?

map() in Pandas

>Works only on a Series

>Applies a function element-wise

>Can also be used for value mapping using a dictionary or Series

>Cannot be used directly on a DataFrame



```
# import pandas as pd

s = pd.Series([1, 2, 3])
print(s.map(lambda x: x * 2))

```

apply() in Pandas

>Works on Series and DataFrames

>Applies a function along an axis (rows or columns)

>More flexible than map()

>Can process multiple values at once



```
# df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

print(df.apply(sum))       # Column-wise
print(df.apply(sum, axis=1))  # Row-wise

```

21. Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy’s array slicing is faster than Python’s list slicing mainly because of how data is stored and how slicing is implemented internally.

>Contiguous Memory Storage

NumPy arrays store elements in a single, contiguous block of memory, while Python lists store references to objects scattered in memory.

>View Instead of Copy

NumPy slicing usually returns a view of the original array (no data is copied), whereas Python list slicing creates a new list and copies elements.

>Low-Level Implementation

NumPy slicing is implemented in optimized C code, while list slicing is handled at the Python level.

>Fixed Data Types

NumPy arrays contain elements of the same data type, allowing efficient memory access and CPU-level optimizations.

>Better Cache Utilization

Contiguous memory improves CPU cache performance, making slicing and subsequent operations faster.

````python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 4, 5, 6]
})

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````
